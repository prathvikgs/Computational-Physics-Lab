# FMD 1

```python

def f(x,y):
    return (-2*y+3)

def explicit_euler(x0,y0,n,h,f):
    x=np.zeros(n)
    y=np.zeros(n)
    x[0]=x0
    y[0]=y0
    for i in range(n-1):
        x[i+1]=x[i]+h
        y[i+1]=y[i]+h*(f(x[i],y[i]))
    
    return x,y    

def implicit_euiler(x0,y0,n,h):
    x=np.zeros(n)
    y=np.zeros(n)
    x[0]=x0
    y[0]=y0
    for i in range (n-1):
        x[i+1]=x[i]+h
        y[i+1]=(y[i]-3*h)/(1+2*h)
    return x,y
```

```python
x1,y1=explicit_euler(0,1,100,0.01,f)
plt.plot(x1,y1)
```

```python
x2,y2=explicit_euler(0,1,100,0.01,f)
plt.plot(x2,y2,'red')
```
