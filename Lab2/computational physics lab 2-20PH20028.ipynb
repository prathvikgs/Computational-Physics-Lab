**a) Code to generate a logistic map given by $X_{n+1}= AX_n(1-X_n)$  with starting value of x as $0 \leq x \leq1$, A is a constant**

```python
def logistic_map(A,ini,iti):
    x=np.zeros(iti)
    n=[]
    n.append(0)
    x[0]=ini
    for i in range(iti-1):
        x[i+1]=A*(x[i]*(1-x[i]))
        n.append(i+1)
    return x,n
    
```

**$0 \leq A \leq1$**

```python
A=0.1
x1,n1=logistic_map(A,0.3,70)
plt.xlabel("No of iterations")
plt.ylabel("xn")
plt.plot(n1,x1)
```

*We observe that the value of xn goes to 0*

**$1 \leq A \leq2$**


```python
A=1.2
x2,n2=logistic_map(A,0.1,70)
plt.xlabel("No of iterations")
plt.ylabel("xn")
plt.plot(n2,x2)
```

*xn approaches (A-1)/A*

**$2 \leq A \leq3$**

```python
A=2.7
x3,n3=logistic_map(A,0.7,50)
plt.xlabel("No of iterations")
plt.ylabel("xn")
plt.plot(n3,x3)
```

*xn approaches (A-1)/A , but initially fluctuates for some time*

**$2 \leq A \leq 1+\sqrt{6}$**

```python
A=3.3
x4,n4=logistic_map(A,0.1,80)
plt.xlabel("No of iterations")
plt.ylabel("xn")
plt.plot(n4,x4)
```

# 1. Method of bisection

```python
def f(x):
    return (2*x**3 -5)
```

```python
def find_root(f,a,b):
    
    if(f(a)*f(b)>0):
        return "No root in the interval"
    
    for i in range(1000):
        c=(a+b)/2
        
        if abs(c-a)<1e-2:
            return c
        
        if(f(a)*f(c))<0:
            b=c
        else:
            a=c
    
```

```python
root1=find_root(f,0,10)
print(root1)
```

```python
find_root(f,-10,root1-0.001)
```

```python
find_root(f,root1+0.001,10)
```

**Recursive Bisection method**

```python
def recursive_root(f,a,b):
    
    if(f(a)*f(b)>0):
        return "No root in the interval "
    
    c=(a+b)/2
    if(abs(c-a)<1e-3):
        return c
    else:
        if(f(a)*f(c)<0):
            return recursive_root(f,a,c)
        else:
            return recursive_root(f,b,c)
```

```python
print(recursive_root(f,-10,20))
```
